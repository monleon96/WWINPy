```python
# 3) Access weight-window values (Block 3 data).
#    The data is structured in wwinp_data.values.particles,
#    where each ParticleBlock has time_bins, energy_bins, w_values.

# Example: Print all w_values for **all** particles, times, energies, geometry
print("\n=== ALL WEIGHT-WINDOW VALUES ===")
for ip, particle_block in enumerate(wwinp_data.values.particles):
    print(f"Particle {ip+1}:")
    for jt, time_slice in enumerate(particle_block.w_values):
        print(f"  Time index: {jt}")
        for ke, geom_values in enumerate(time_slice):
            print(f"    Energy index: {ke}")
            print(f"      w_values: {geom_values}")
```

```python
# 4) Print all w_values for a certain energy bin (example: ip=0, energy bin=2)
#    Check bounds carefully to avoid IndexError if the bin does not exist.
print("\n=== W_VALUES FOR A SPECIFIC ENERGY BIN (PARTICLE=1, ENERGY BIN=2) ===")
particle_index = 0
energy_bin_index = 1  # i.e. "2nd bin" if zero-based
if (particle_index < len(wwinp_data.values.particles) and
    energy_bin_index < len(wwinp_data.values.particles[particle_index].energy_bins)):
    particle_block = wwinp_data.values.particles[particle_index]
    # We iterate over time indices
    for t_index, time_slice in enumerate(particle_block.w_values):
        if energy_bin_index < len(time_slice):
            geom_values = time_slice[energy_bin_index]
            print(f"Time index: {t_index}, E bin: {energy_bin_index}")
            print(f"w_values: {geom_values}")
else:
    print("Requested particle or energy bin index is out of range.")
```

```python
# 5) Print all w_values for a certain energy bin and a fixed z
#    We assume you have a 3D mesh and want to slice it along z.
#    Well define a helper to flatten geometry indices into (ix, iy, iz).
#    This assumes your parser stores geometry in a single 1D list per (time, energy).
print("\n=== W_VALUES FOR E BIN=2 AND FIXED Z=2 (PARTICLE=1) ===")
# Example approach:
def flatten_index(ix, iy, iz, nx, ny, nz):
    """
    Convert (ix, iy, iz) 3D indices into a single 1D index
    for an array shaped [nx, ny, nz].
    """
    return iz + nz * (iy + ny * ix)

# For demonstration, let's pick z=2 as the "fixed Z plane".
# We need to figure out how many fine cells are in the z direction.
# We'll compute total_x_fine, total_y_fine, total_z_fine:
total_x_fine = sum(seg.s for seg in wwinp_data.geometry.x_axis.segments)
total_y_fine = sum(seg.s for seg in wwinp_data.geometry.y_axis.segments)
total_z_fine = sum(seg.s for seg in wwinp_data.geometry.z_axis.segments)

# Convert them to int if s was float in the input
nx = int(total_x_fine)
ny = int(total_y_fine)
nz = int(total_z_fine)

z_plane = 2  # zero-based index for the cell in z
if z_plane < nz:
    # Retrieve the w_values for the same bin used above (particle=0, energy_bin_index=1)
    pb = wwinp_data.values.particles[particle_index]
    for t_index, time_slice in enumerate(pb.w_values):
        if energy_bin_index < len(time_slice):
            geom_vals_1d = time_slice[energy_bin_index]
            # Now print only the slice at z_plane:
            row_values = []
            for ix in range(nx):
                for iy in range(ny):
                    idx_1d = flatten_index(ix, iy, z_plane, nx, ny, nz)
                    w_val = geom_vals_1d[idx_1d]
                    row_values.append((ix, iy, z_plane, w_val))
            print(f"Time index = {t_index}, E bin = {energy_bin_index}")
            print("  (ix, iy, iz, w_val) for z=2 plane:")
            for rv in row_values:
                print(f"    {rv}")
else:
    print(f"Requested z={z_plane} is out of range (nz={nz}).")
```

```python
# 6) Print the w_value of a specific cell, e.g. (ix=1, iy=2, iz=3) for the same bin
print("\n=== W_VALUE OF A SPECIFIC CELL (ix=1, iy=2, iz=3) ===")
ix, iy, iz = 1, 2, 3
if ix < nx and iy < ny and iz < nz:
    idx_1d = flatten_index(ix, iy, iz, nx, ny, nz)
    pb = wwinp_data.values.particles[particle_index]
    # Just grab the first time slice for demonstration
    if len(pb.w_values) > 0:
        time_slice = pb.w_values[0]
        if energy_bin_index < len(time_slice):
            geom_vals_1d = time_slice[energy_bin_index]
            if idx_1d < len(geom_vals_1d):
                print(f"w_value at (ix={ix}, iy={iy}, iz={iz}): {geom_vals_1d[idx_1d]}")
            else:
                print("Index in geometry array is out of range.")
        else:
            print("Energy bin index is out of range in the time slice.")
    else:
        print("No time slices available for this particle.")
else:
    print("Requested cell indices are out of range.")
```

```python
# 7) Print the geometry in x, y, and z
#    For example, print the origin and the coarse mesh segments along x, y, z.
print("\n=== GEOMETRY (X, Y, Z) ===")
geom = wwinp_data.geometry
print(f"X-axis origin = {geom.x_axis.origin}")
print("  Coarse segments (q, p, s):")
for seg in geom.x_axis.segments:
    print(f"    q={seg.q}, p={seg.p}, s={seg.s}")

print(f"Y-axis origin = {geom.y_axis.origin}")
print("  Coarse segments (q, p, s):")
for seg in geom.y_axis.segments:
    print(f"    q={seg.q}, p={seg.p}, s={seg.s}")

print(f"Z-axis origin = {geom.z_axis.origin}")
print("  Coarse segments (q, p, s):")
for seg in geom.z_axis.segments:
    print(f"    q={seg.q}, p={seg.p}, s={seg.s}")
```

```python
# 8) Print only the coarse mesh geometry of x axis
print("\n=== COARSE MESH GEOMETRY OF X AXIS ===")
print(f"X-axis origin = {geom.x_axis.origin}")
for i, seg in enumerate(geom.x_axis.segments, start=1):
    print(f"  Segment {i}: q={seg.q}, p={seg.p}, s={seg.s}")
```

```python
# 9) Print coarse and fine mesh geometry of y axis
#    Fine mesh geometry often requires you to compute the boundaries
#    of each fine mesh. Well do a quick example:
print("\n=== COARSE AND FINE MESH GEOMETRY OF Y AXIS ===")
print(f"Y-axis origin = {geom.y_axis.origin}")
coarse_end = geom.y_axis.origin
for i, seg in enumerate(geom.y_axis.segments, start=1):
    coarse_start = coarse_end
    coarse_end = seg.p  # p is the end coordinate of this coarse segment
    print(f"  Coarse segment {i}: [start={coarse_start}, end={coarse_end}], "
            f"q={seg.q}, s={seg.s}")
    if seg.s > 0:
        fine_delta = (coarse_end - coarse_start) / seg.s
        fine_boundaries = [coarse_start + n * fine_delta for n in range(int(seg.s) + 1)]
        print(f"    Fine mesh boundaries: {fine_boundaries}")
```
